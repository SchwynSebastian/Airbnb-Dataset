## 2.3 Graphik

```python
import matplotlib.pyplot as plt


minprice = Airbnb['price'].min()
maxprice = Airbnb['price'].max()

# Preisintervall (Bins)
bins = [0, 50, 100, 150, 200, 300, 400, 500, maxprice]  # Preiskategorien
labels = ['0-50', '51-100', '101-150', '151-200', '201-300', '301-400', '401-500', f'{500}+']  # Kategorienbeschriftungen

# Kategorisieren der Daten
Airbnb['price_category'] = pd.cut(Airbnb['price'], bins=bins, labels=labels)

# Häufigkeitder Preisgruppen zählen
price_count = Airbnb['price_category'].value_counts().sort_index()

plt.figure(figsize=(10,6))
price_count.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Verteilung der Preise (Säulendiagramm)', fontsize=14)
plt.xlabel('Preiskategorien', fontsize=12)
plt.ylabel('Häufigkeit', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

## 2.4 Skalieren

Ich finde, dass eine Skalierung in meinem Fall nicht sinnvoll ist, weil die Preisverteilung so schon gut erkennbar ist. Eine Skalierung würde die kleineren Werte zu stark hervorheben und die Darstellung unübersichtlicher machen. Außerdem ist eine normale Darstellung einfacher zu verstehen, weil sie die tatsächlichen Unterschiede zwischen den Preiskategorien klar zeigt. Gerade bei der Analyse von Preisen ist es wichtig, dass die Verteilung unverändert bleibt, damit keine falschen Eindrücke entstehen.
