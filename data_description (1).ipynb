## 2.3 Graphik

```python
import matplotlib.pyplot as plt


minprice = Airbnb['price'].min()
maxprice = Airbnb['price'].max()

# Preisintervall (Bins)
bins = [0, 50, 100, 150, 200, 300, 400, 500, maxprice]  # Preiskategorien
labels = ['0-50', '51-100', '101-150', '151-200', '201-300', '301-400', '401-500', f'{500}+']  # Kategorienbeschriftungen

# Kategorisieren der Daten
Airbnb['price_category'] = pd.cut(Airbnb['price'], bins=bins, labels=labels)

# HÃ¤ufigkeitder Preisgruppen zÃ¤hlen
price_count = Airbnb['price_category'].value_counts().sort_index()

plt.figure(figsize=(10,6))
price_count.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Verteilung der Preise (SÃ¤ulendiagramm)', fontsize=14)
plt.xlabel('Preiskategorien', fontsize=12)
plt.ylabel('HÃ¤ufigkeit', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

## 2.4 Skalieren

Ich finde, dass eine Skalierung in meinem Fall nicht sinnvoll ist, weil die Preisverteilung so schon gut erkennbar ist. Eine Skalierung wÃ¼rde die kleineren Werte zu stark hervorheben und die Darstellung unÃ¼bersichtlicher machen. AuÃerdem ist eine normale Darstellung einfacher zu verstehen, weil sie die tatsÃ¤chlichen Unterschiede zwischen den Preiskategorien klar zeigt. Gerade bei der Analyse von Preisen ist es wichtig, dass die Verteilung unverÃ¤ndert bleibt, damit keine falschen EindrÃ¼cke entstehen. **Hier wÃ¤re Beispielsweise noch eine Skalierung (als beispiel) fÃ¼r den Preis:**

```python
price = Airbnb["price"]
Airbnb["price_normalised"] = (price - price.min()) / (price.max() - price.min())
Airbnb["price_normalised"].describe()
```
